```python
import pandas as pd
import geopandas as gpd
```

```python
harrogate = gpd.read_file('harrogate.shp')
st_helen = gpd.read_file('st_helen.shp')
worksop = gpd.read_file('worksop.shp')
```

```python
harrogate.head(2)
```

```python
harrogate.columns
```

```python
# Select relevant columns
keep_cols = [
    "geometry", 
    "Catchmen_1", 
    "Dominant S", 
    "Catchmen_2"
]
```

```python
harrogate = harrogate[keep_cols]

# Rename for consistency
harrogate = harrogate.rename(columns={
    "Catchmen_1": "Catchment_Name",
    "Dominant S": "Dominant_Segment",
    "Catchmen_2": "Catchment_Category"
})

# Add station name column
harrogate["Station"] = "Harrogate"

harrogate.head()
```

```python
st_helen = st_helen[keep_cols]

# Rename for consistency
st_helen = st_helen.rename(columns={
    "Catchmen_1": "Catchment_Name",
    "Dominant S": "Dominant_Segment",
    "Catchmen_2": "Catchment_Category"
})

# Add station name column
st_helen["Station"] = "st_helen"

st_helen.head()
```

```python
worksop = worksop[keep_cols]

# Rename for consistency
worksop = worksop.rename(columns={
    "Catchmen_1": "Catchment_Name",
    "Dominant S": "Dominant_Segment",
    "Catchmen_2": "Catchment_Category"
})

# Add station name column
worksop["Station"] = "worksop"

worksop.head()
```

```python
# Merge all three station GeoDataFrames
all_catchments = pd.concat([harrogate, st_helen, worksop], ignore_index=True)
```

```python
all_catchments.head()
```

```python
all_catchments['Catchment_Category'].value_counts()
```

```python
all_catchments.to_file("catchments.geojson", driver="GeoJSON")
```
